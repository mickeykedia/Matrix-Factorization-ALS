```python
ratings = sc.textFile("/Users/mayankkedia/code/distributed_computing/movies/ratings.txt")
```

```python
ratings = ratings.map(lambda x: x.split(":")[0:3])
# Asking spark to cache the RDD into memory once it has been split
ratings.map(lambda x: (int(x[0], int(x[1], float(x[2])))))
ratings.cache()
```

```python
user_ratings = ratings.map(lambda x:(x[0], x[1:3]))
```

```python
user_ratings = user_ratings.groupByKey()
```

```python
# user_ratings.persist(Memory and disk option)
# GroupByKey hash partitions the RDD anyway
user_ratings.take(1)
```

```python
user_ratings.getNumPartitions()
```

```python
## Calling partitionBy() with a hashpartitioner apparently tells spark that there has been 
# an explicit partitioning by key and then later join operations will work quite well 
```

```python
item_ratings = ratings.map(lambda x:(x[1], (x[0], x[2]))).groupByKey()
# item_ratings.persist(Memory and disk option)
#
```
